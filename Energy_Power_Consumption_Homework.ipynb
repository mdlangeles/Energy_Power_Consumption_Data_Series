# M√©tricas de Evaluaci√≥n de Modelos Predictivos

Las principales m√©tricas de evaluaci√≥n de predicciones que se pueden usar para crear y probar modelos predictivos de valores num√©ricos:

## 1. Error Medio (ME ‚Äì Mean Error)
El Error Medio (ME) es el promedio de los errores en un conjunto de observaciones. Aunque es una m√©trica simple y f√°cil de interpretar, sufre de sesgo debido a la compensaci√≥n entre errores positivos y negativos, lo que puede ocultar imprecisiones en predicciones individuales. Sin embargo, es √∫til para analizar la simetr√≠a en la distribuci√≥n de errores y detectar posibles sesgos en el modelo.

$$
ME = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)
$$

---

## 2. Error Absoluto Medio (MAE ‚Äì Mean Absolute Error)
El Error Absoluto Medio (MAE) es una soluci√≥n al problema de la cancelaci√≥n de errores del ME, ya que usa valores absolutos. Proporciona el promedio de la magnitud de los errores sin importar su signo, permitiendo una evaluaci√≥n m√°s precisa de la inexactitud en las predicciones.

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

---

## 3. Error Cuadr√°tico Medio (MSE ‚Äì Mean Square Error)
El Error Cuadr√°tico Medio (MSE), al igual que el MAE, evita la cancelaci√≥n de errores positivos y negativos. Sin embargo, penaliza m√°s los errores grandes, lo que lo hace √∫til cuando se quiere dar mayor importancia a desviaciones significativas en las predicciones.

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

---

## 4. Ra√≠z del Error Cuadr√°tico Medio (RMSE ‚Äì Root Mean Square Error)
La Ra√≠z del Error Cuadr√°tico Medio (RMSE) es la desviaci√≥n est√°ndar de los errores. Comparte las ventajas del MSE, pero al expresarse en las mismas unidades que la variable predicha, facilita la interpretaci√≥n. Es ampliamente utilizada en predicciones y an√°lisis de regresi√≥n para evaluar la precisi√≥n de los modelos.

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

---

## 5. Error Porcentual Medio (MPE ‚Äì Mean Percentage Error)
El Error Porcentual Medio (MPE) mide el error promedio en porcentaje respecto a los valores reales. Es f√°cil de interpretar y √∫til para detectar sesgos en las predicciones. Sin embargo, sufre de compensaci√≥n entre errores positivos y negativos y no es adecuado cuando hay valores observados iguales a cero, ya que podr√≠a generar resultados indefinidos.

$$
MPE = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{y_i - \hat{y}_i}{y_i} \times 100\right)
$$

---

## 6. Error Porcentual Absoluto Medio (MAPE ‚Äì Mean Absolute Percentage Error)
El Error Porcentual Absoluto Medio (MAPE) soluciona el problema de la compensaci√≥n de errores al usar valores absolutos, como el MAE. Funciona bien cuando no hay valores extremos ni ceros en los datos, ya que estos pueden distorsionar los resultados.

$$
MAPE = \frac{1}{n} \sum_{i=1}^{n} \left|\frac{y_i - \hat{y}_i}{y_i} \times 100\right|
$$

---

## 7. Criterio de Informaci√≥n de Akaike (AIC)
Eval√∫a la calidad de un modelo estad√≠stico considerando tanto el ajuste como la complejidad. Un AIC m√°s bajo indica un mejor equilibrio entre precisi√≥n y simplicidad. Se utiliza para comparar modelos diferentes; el modelo con el AIC m√°s bajo es preferible. Es especialmente √∫til en la selecci√≥n de modelos cuando se busca evitar el sobreajuste.

$$
AIC = 2k - 2 \ln(L)
$$

donde:
- \( k \) es el n√∫mero de par√°metros del modelo,
- \( L \) es la verosimilitud del modelo.

---

## 8. Criterio de Informaci√≥n Bayesiano (BIC)
Similar al AIC, pero penaliza m√°s la complejidad del modelo, especialmente cuando el tama√±o de la muestra es grande. Un BIC m√°s bajo sugiere un modelo m√°s parsimonioso. Adecuado cuando se dispone de grandes conjuntos de datos y se desea evitar modelos excesivamente complejos.

$$
BIC = k \ln(n) - 2 \ln(L)
$$

donde:
- \( n \) es el n√∫mero de observaciones,
- \( k \) es el n√∫mero de par√°metros del modelo,
- \( L \) es la verosimilitud del modelo.

# **SARIMAX**

```python
# Definir los par√°metros p, d, q para la parte no estacional
p = d = q = range(0, 3)
pdq = list(itertools.product(p, d, q))

# Definir los par√°metros P, D, Q para la parte estacional (con periodicidad de 12)
P = D = Q = range(0, 2)
seasonal_pdq = [(x, y, z, 12) for x in P for y in D for z in Q]  # Periodo = 12

# Variables ex√≥genas m√°s correlacionadas con PowerConsumption_Zone2
exog_vars = ["PowerConsumption_Zone1", "PowerConsumption_Zone3"]
exog_train = data_scaled[exog_vars][:date_train]
exog_test = data_scaled[exog_vars][date_train:]

best_score = float("inf")
best_param = None
best_seasonal_param = None
```

```python
# B√∫squeda del mejor modelo SARIMAX
print("Buscando el mejor modelo SARIMAX...")
st = time.time()
for param in pdq:
    for seasonal_param in seasonal_pdq:
        try:
            model = sm.tsa.SARIMAX(y_train, order=param, seasonal_order=seasonal_param, exog=exog_train)
            results = model.fit()
            if results.aic < best_score:
                best_score = results.aic
                best_param = param
                best_seasonal_param = seasonal_param
        except:
            continue
et = time.time()

print(f"Mejor Par√°metro SARIMAX: {best_param} x {best_seasonal_param}")
print(f"Tiempo de ejecuci√≥n: {et - st} segundos")
```

```python
# Entrenar el mejor modelo SARIMAX
best_sarimax = sm.tsa.SARIMAX(y_train, order=best_param, seasonal_order=best_seasonal_param, exog=exog_train)
results_sarimax = best_sarimax.fit()

# Predicci√≥n con SARIMAX
pred_uc_sarimax = results_sarimax.get_forecast(steps=len(y_test), exog=exog_test)
pred_ci_sarimax = pred_uc_sarimax.conf_int()
```

```python
# Evaluaci√≥n del modelo
y_pred_sarimax = pred_uc_sarimax.predicted_mean
rmse_sarimax = sqrt(metrics.mean_squared_error(y_test, y_pred_sarimax))
print(f"RMSE SARIMAX: {rmse_sarimax}")
```

# **M√âTRICAS DE EVALUACI√ìN M√ÅS ADECUADAS A UTILIZAR**

Seg√∫n nuestro criterio de evaluaci√≥n, e investigaci√≥n sobre m√©tricas, hemos tomado la ecisi√≥n de usar las siguientes para nuestro an√°lisis:

### 1Ô∏è‚É£ Error Absoluto Medio (MAE)
- √ötil porque proporciona una medida clara y f√°cil de interpretar del error promedio en las mismas unidades del consumo energ√©tico.  
- Es ideal cuando queremos entender el error t√≠pico sin preocuparnos demasiado por valores at√≠picos.  

### 2Ô∏è‚É£ Ra√≠z del Error Cuadr√°tico Medio (RMSE)
- Similar al MAE, pero m√°s sensible a los errores grandes, lo que ayuda a detectar momentos en los que el modelo falla gravemente en la predicci√≥n.  
- Dado que el consumo energ√©tico puede tener picos altos, el RMSE nos ayudar√° a evaluar si el modelo maneja bien esas variaciones.  




```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import numpy as np

# Funci√≥n para calcular las m√©tricas
def calcular_metricas(y_real, y_pred):
    mae = mean_absolute_error(y_real, y_pred)
    rmse = np.sqrt(mean_squared_error(y_real, y_pred))
    mape = np.mean(np.abs((y_real - y_pred) / y_real)) * 100  # Expresado en porcentaje
    return mae, rmse, mape

# Calcular m√©tricas para ARIMA
mae_arima, rmse_arima, mape_arima = calcular_metricas(y_test, pred_uc.predicted_mean)
print(f"ARIMA -> MAE: {mae_arima:.2f}, RMSE: {rmse_arima:.2f}")

# Calcular m√©tricas para SARIMA
mae_sarima, rmse_sarima, mape_sarima = calcular_metricas(y_test, pred_uc_s.predicted_mean)
print(f"SARIMA -> MAE: {mae_sarima:.2f}, RMSE: {rmse_sarima:.2f}")

# Calcular m√©tricas para ARIMAX
mae_arimax, rmse_arimax, mape_arimax = calcular_metricas(y_test, pred_uc_x.predicted_mean)
print(f"ARIMAX -> MAE: {mae_arimax:.2f}, RMSE: {rmse_arimax:.2f}")

# Calcular m√©tricas para SARIMAX
mae_sarimax, rmse_sarimax, mape_sarimax = calcular_metricas(y_test, pred_uc_sarimax.predicted_mean)
print(f"SARIMAX -> MAE: {mae_sarimax:.2f}, RMSE: {rmse_sarimax:.2f}")
```

Analizando las m√©tricas de error, el modelo ARIMAX presenta el menor MAE (0.24) y RMSE (0.28), lo que indica que su error medio y la penalizaci√≥n de valores extremos son m√°s bajos en comparaci√≥n con los dem√°s modelos. SARIMAX, aunque cercano en desempe√±o (MAE: 0.25, RMSE: 0.32), tiene un error ligeramente mayor. En contraste, ARIMA y SARIMA muestran errores m√°s altos, especialmente SARIMA (MAE: 0.56, RMSE: 0.73), lo que sugiere que sin variables ex√≥genas, estos modelos no logran capturar adecuadamente la din√°mica del consumo energ√©tico. Adem√°s, en las gr√°ficas de predicci√≥n, se observa que ARIMAX y SARIMAX generan ajustes m√°s precisos con menor dispersi√≥n en los intervalos de confianza, lo que refuerza la efectividad de incluir variables ex√≥genas.
